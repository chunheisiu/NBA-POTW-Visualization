```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/National_Basketball_Association'
```

```python
raw_table = pd.read_html(url)[2]
cols = [a[0] for a in raw_table.columns]
raw_table.columns = cols
```

```python
df = raw_table.drop(index = 15).reset_index(drop = True).drop(columns = 'Unnamed: 8_level_0')
```

```python
cood = df["Coordinates"].str.split("/", n = 1, expand = True) 
df["Coordinates"] = cood[1]
```

```python
cood = df["Coordinates"].str.split("°N ", n = 1, expand = True) 
cood2 = cood[1].str.split("°W", n = 1, expand = True)
cood[1] = cood2[0]
cood.columns = ['latitude','longitude']
df_new = pd.concat([df, cood], axis = 1)
df_new
```

```python
df_new.to_csv('../data/scraped/NBA_Location')
```
